```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('sentence-transformers/all-MiniLM-L12-v2') 
```

```python
file_path = 'test_news.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    document = file.read()

sentences_a = document.split('\n')
```

```python
len(sentences_a)
```

```python
sentences_b = []
current_subset = ""
for i, sentence in enumerate(sentences_a):
    current_subset += sentence + '. '
    sentences_b.append(current_subset)
```

```python
sentences_b
```

```python
len(sentences_b)
```

```python
ebd_a = model.encode(sentences_a)
ebd_b = model.encode(sentences_b)
```

```python
# ebd_b i번째, ebd_a i+1번째 비교
```

```python
similarities = []
```

```python
for i in range(41):
    similarity_i = cosine_similarity([ebd_b[i]], [ebd_a[i + 1]])[0][0]
    similarities.append(similarity_i)
```

```python
len(similarities)
```

```python
plt.plot(range(0, 41), similarities, linestyle='-', color='black', label='Cosine Similarity')
plt.xlabel('Index')
plt.ylabel('Cosine Similarity')
plt.legend()
plt.show()
```
